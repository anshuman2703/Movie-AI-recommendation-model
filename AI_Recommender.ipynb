```python
import pandas as pd

df = pd.read_csv("movies.csv")
```

```python
def recommend_from_preferences(liked_titles, df):
    liked_titles = [t.strip() for t in liked_titles]
    
    liked_genres = df[df['Title'].isin(liked_titles)]['Genre']
    liked_types = df[df['Title'].isin(liked_titles)]['Type']
    
    recommendations = df[
        (df['Genre'].isin(liked_genres)) &
        (df['Type'].isin(liked_types)) &
        (~df['Title'].isin(liked_titles))
    ]
    
    recommendations = recommendations.sort_values(by='Rating', ascending=False)
    
    return recommendations[['Title','Type','Genre','Rating']]
```

```python
user_input = input("Enter movies, series or anime that you like: ")

liked = user_input.split(",")

results = recommend_from_preferences(liked, df)

print("\nAI Recommendations based on your taste:\n")
print(results)
```
